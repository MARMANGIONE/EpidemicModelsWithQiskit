```python
from qiskit import *
q = QuantumRegister(5,"q")
c = ClassicalRegister(5,"c")
circuit = QuantumCircuit(q,c)


circuit.h(q[0])
circuit.h(q[1])
circuit.h(q[2])
circuit.h(q[3])
circuit.x(q[4])
circuit.u1(-0.07390596717, q[0])
circuit.u1(-0.13665928043, q[1])
circuit.u1(-0.20765927440, q[2])
circuit.u1(-0.17514379043, q[3])
circuit.cx(q[4],q[0])
circuit.u1(0.1114479993917, q[0])
circuit.cx(q[4],q[1])
circuit.u1(0.27174776453, q[1])
circuit.h(q[0])
circuit.cx(q[4],q[2])
circuit.u1(0.8573406351, q[2])
circuit.h(q[1])
circuit.cx(q[4],q[3])
circuit.u1(2.96644886315, q[3])
circuit.h(q[2])
circuit.h(q[3])
circuit.measure(q[3], c[3])
circuit.measure(q[2], c[2])
circuit.measure(q[1], c[1])
circuit.measure(q[0], c[0])
circuit.measure(q[4], c[4])
```

```python
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend = simulator).result()
from qiskit.tools.visualization import plot_histogram
plot_histogram(result.get_counts(circuit))
```
