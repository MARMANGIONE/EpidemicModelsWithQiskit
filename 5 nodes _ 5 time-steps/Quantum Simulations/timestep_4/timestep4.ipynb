```python
from qiskit import *
q = QuantumRegister(5,"q")
c = ClassicalRegister(5,"c")
circuit = QuantumCircuit(q,c)


circuit.h(q[0])
circuit.h(q[1])
circuit.h(q[2])
circuit.h(q[3])
circuit.x(q[4])
circuit.u1(-0.0781471, q[0])
circuit.u1(-0.149383, q[1])
circuit.u1(-0.229336, q[2])
circuit.u1(-0.106814, q[3])
circuit.cx(q[4],q[0])
circuit.u1(-0.0624392, q[0])
circuit.cx(q[4],q[1])
circuit.u1(0.00455531, q[1])
circuit.h(q[0])
circuit.cx(q[4],q[2])
circuit.u1(-0.276146, q[2])
circuit.h(q[1])
circuit.cx(q[4],q[3])
circuit.u1(2.50071, q[3])
circuit.h(q[2])
circuit.h(q[3])
circuit.measure(q[3], c[3])
circuit.measure(q[2], c[2])
circuit.measure(q[1], c[1])
circuit.measure(q[0], c[0])
circuit.measure(q[4], c[4])
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend = simulator).result()
from qiskit.tools.visualization import plot_histogram
plot_histogram(result.get_counts(circuit))
```
