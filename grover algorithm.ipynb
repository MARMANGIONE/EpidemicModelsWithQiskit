```python
#initialization
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
%config InlineBackend.figure_format = 'svg' # Makes the images look nice

# importing Qiskit
from qiskit import IBMQ, Aer
from qiskit.providers.ibmq import least_busy
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister, execute

# import basic plot tools
from qiskit.visualization import plot_histogram
```

```python
n = 2
q = QuantumRegister(n)
c = ClassicalRegister(n)
grover_circuit = QuantumCircuit(q,c)
```

```python
for qubit in range(n):
    grover_circuit.h(qubit)
grover_circuit.draw('mpl')
```

```python
for i in range(n):
    grover_circuit.s(q[i])
grover_circuit.cz(q[0], q[1])
for i in range(n):
    grover_circuit.s(q[i])
grover_circuit.draw('mpl')
```

```python
for i in range(n):
    grover_circuit.h(q[i])

for i in range(n):
    grover_circuit.x(q[i])

grover_circuit.cz(q[0], q[1])

for i in range(n):
    grover_circuit.x(q[i])

for i in range(n):
    grover_circuit.h(q[i])
grover_circuit.draw('mpl')
```

```python
grover_circuit.measure(q, c)
grover_circuit.draw('mpl')
```

```python
backend_sim = Aer.get_backend('statevector_simulator')
job_sim = execute(grover_circuit, backend_sim)
statevec = job_sim.result().get_statevector()
grover_circuit.measure_all()
backend = Aer.get_backend('qasm_simulator')
shots = 1024
results = execute(grover_circuit, backend=backend, shots=shots).result()
answer = results.get_counts()
plot_histogram(answer)
```
